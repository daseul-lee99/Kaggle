#### The data

* **Data** 탭을 클릭해서 아래로 내려보면 데이터를 확인할 수 있음
* **Data Sources* 아래에 파일을 클릭하면 오른쪽 **Clumns**에서 데이터의 모든 컬럼을 볼 수 있음
* 그 아래에서 더 상세한 데이터 정보를 볼 수 있음
1. train.csv
    * 승객 데이터 891개(891명)
    * 두번째 컬럼의 **Survived**: 1이면 생존, 0이면 죽음
    * 첫번째 승객인 Mr. Owen Harris Braund은 22살이고, 죽었다는 것을 알 수 있음
2. test.csv
    * 승객 데이터 418개(418명)
    * Survived 컬럼이 없음 -> 내가 이 값을 얼마나 잘 예측하는지에 따라 높은 점수를 얻을 수 있음
3. gender_submission.csv
    * 내가 한 예측을 어떻게 구조화 해야하는지 보여주는 파일
    * **PassengerId": test.csv의 승객 ID 값
    * **Survived**: 승객이 살았으면 1, 죽었으면 0으로 표현

## Part 3: Improve your score

* goal: test.csv에 있는 승객들의 생사여부를 예측하는 패턴을 train.csv에서 찾는 것

#### Explore a pattern

* gender_submission.csv: 모든 여자 승객은 생존, 남자 승객은 사망으로 가정한 결과
* 위 패턴이 실제인지 train data로 확인해보기

<br>

* 성별이 여자(female)인 사람의 생존 비율

<img src="./images/WomenRate.png" width="70%">

* 성별이 남자(male)인 사람의 생존 비율

<img src="./images/MenRate.png" width="70%">

* 위 결과에서 성별이 생존여부에 있어 중요한 요인이라는 것을 알 수 있으므로, gender_submission.csv는 나쁘지 않은 추측이라 할 수 있음
* But, gender_submission.csv는 성별 컬럼 한 개만 이용하고 있음
* 여러개의 컬럼을 활용하면 더 복잡한(더 나은 예측을 하는) 패턴을 찾을 수 있음
* 한 번에 여러개의 컬럼을 고려하는 것은 꽤 어렵기 때문에(or 수 많은 컬럼에서 가능한 모든 패턴을 고려하는 것은 시간이 꽤 걸리기 때문에), 머신 러닝을 활용해 이 과정을 자동화 함

<br>
