## Interpreting Data Description

* count: 값이 모두 들어있는 행의 수
    * Bedroom2 컬럼의 경우, 화장실이 1개만 있는 집이면 값이 없음(실제 csv 파일에서는 비어있으나 위에서는 0으로 출력됨)
* mean: 컬럼의 평균값
* std: 컬럼의 표준편차

컬럼을 오름차순 정렬했을 때,
* min: 첫번째 값(제일 작은 값)
* 25/50/75%: (정렬된 컬럼의 전체 개수에서)20/50/75% 위치의 값(데이터가 100개면 25/50/75번째)
* max: 마지막 값(제일 큰 값)
